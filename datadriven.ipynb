```python
#インストールしたpackageをimport

import matplotlib.pyplot as plt
import numpy as np
#matplotlib inline
from matplotlib import animation
from IPython.display import HTML

def init():
    line.set_data([],[])
    return(line,)

def animate(i):
    state = state_history[i][0]
    x = (state % 3) + 0.5
    y = 2.5 - int(state / 3)
    line.set_data(x,y)
    return(line,)


fig = plt.figure(figsize=(5,5))
ax= plt.gca()
color= "red"

plt.plot([1,1],[0,1],color, linewidth=2)
plt.plot([1,2],[2,2],color, linewidth=2)
plt.plot([2,2],[2,1],color, linewidth=2)
plt.plot([2,3],[1,1],color, linewidth=2)

plt.text(0.5, 2.5, "S0", size=14, ha="center")
plt.text(1.5, 2.5, "S1", size=14, ha="center")
plt.text(2.5, 2.5, "S2", size=14, ha="center")
plt.text(0.5, 1.5, "S3", size=14, ha="center")
plt.text(1.5, 1.5, "S4", size=14, ha="center")
plt.text(2.5, 1.5, "S5", size=14, ha="center")
plt.text(0.5, 0.5, "S6", size=14, ha="center")
plt.text(1.5, 0.5, "S7", size=14, ha="center")
plt.text(2.5, 0.5, "S8", size=14, ha="center")
plt.text(0.5, 2.3, "START", ha="center")
plt.text(2.5, 0.3, "GOAL", ha="center")

ax.set_xlim(0, 3)
ax.set_ylim(0, 3)

#描画範囲の設定とメモリを消す
plt.tick_params(axis="both", which="both", bottom="off", top="off",
               labelbottom="off", right="off", left="off", labelleft="off")

#現在地S0に緑丸を描画する
line, = ax.plot([0.5],[2.5],marker="o",color="g",markersize=60)

#行は状態0~7、列は移動方向で ↑→↓←を表す
theta_0 = np.array([
    [np.nan, 1, 1, np.nan], #s0
    [np.nan, 1, np.nan, 1], #s1
    [np.nan, np.nan, 1, 1], #s2
    [1, 1, 1, np.nan], #s3
    [np.nan, np.nan, 1, 1], #s4
    [1, np.nan, np.nan, np.nan], #s5
    [1, np.nan, np.nan, np.nan], #s6
    [1, 1, np.nan, np.nan,], #s7
])




def softmax(theta):
    beta = 1
    
    [m,n]=theta.shape #thetaの行列サイズ
    
    pi = np.zeros((m,n))
    
    exp_theta = np.exp(beta * theta)
    
    for i in range(0, m):
        pi[i, :] = exp_theta[i, :] / np.nansum(exp_theta[i,:]) #割合の計算
    pi = np.nan_to_num(pi) #nanを０に変換
    return pi


pi_0 = softmax(theta_0)



#theta_0と、piとstatehistoryが入る
def update_theta(theta, pi, state_history):
    eta = 0.1
    T = len(state_history)
    
    [m,n] = theta.shape
    delta_theta = theta.copy()
    for i in range(0, m):
        for j in range(0,n):
            if not(np.isnan(theta[i,j])):
                SA_i = [SA for SA in state_history if SA[0] == i] #状態０～８
                
                SA_ij = [SA for SA in state_history if SA ==[i, j]] #状態0の時、0～3を取ったもの
                
                N_i = len(SA_i)
                N_ij = len(SA_ij)
                delta_theta[i,j] = (N_ij + pi[i, j] * N_i) / T #各行と列に式の値を代入
                
    new_theta = theta + eta * delta_theta
    return new_theta




def get_next_s(pi, s):
    direction = ["up", "right", "down", "left"]
    next_direction = np.random.choice(direction, p = pi[s,:])
    #pi[s, :]の確率にしたがって、directionが選択される。
    
    if next_direction == "up":
        action = 0
        s_next = s - 3
        
    elif next_direction == "right":
        action = 1
        s_next = s + 1
        #右rに移動するときは状態の数字が1大きくなる
        
    elif next_direction == "down":
        action = 2
        s_next = s + 3 
        #下に移動する時は状態の数字が3大きくなる
        
    elif next_direction == "left":
        action = 3
        s_next = s-1
        
    return [action, s_next] #次の状態（現在地）と、取った行動を返す　SとA



def goal_maze(pi):
    s = 0 #スタート地点
    state_history = [[0, np.nan]]
    
    while(1):
        [action, next_s] = get_next_s(pi, s)
        state_history[-1][1] = action #前の状態に、アクションを追加　/状態と取ったアクションのセットができる/
        state_history.append([next_s, np.nan])#次の状態と、未定のアクションnanを追加
        
        if next_s == 8:
            break
        else:
            s = next_s
    return state_history



#方策勾配法で迷路を解く

stop_epsilon = 10**-8

theta = theta_0
pi = pi_0

is_continue = True
count = 1

while is_continue:
    state_history = goal_maze(pi) #方策πで迷路内を探索した履歴を求める
    new_theta = update_theta(theta, pi, state_history) #パラメータΘを更新
    new_pi = softmax(new_theta) #方策πの更新
    #print(np.sum(np.abs(new_pi - pi)))
    #print("迷路を解くのにかかったステップ"+str(len(state_history)-1))
    
    if np.sum(np.abs(new_pi - pi)) < stop_epsilon:
        is_continue = False
    else:
        theta = new_theta
        pi = new_pi

np.set_printoptions(precision=3, suppress=True)
print(pi)
        
anim = animation.FuncAnimation(fig, animate, init_func=init, frames=len(state_history), interval = 200, repeat=False)
HTML(anim.to_jshtml())
```
